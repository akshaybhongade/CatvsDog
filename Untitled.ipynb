```python
import numpy as np
#import tensorflow as tf
from zipfile import ZipFile
import matplotlib.pyplot as plt
from matplotlib import image
from PIL import Image
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
```

```python
filename  = "Cat_vs_Dog_Dataset.zip"
with ZipFile(filename) as catdog:
    catdog.extractall()
        
```

```python

def createdata(test_percent,im_width, im_height):

    width= im_width
    hight =  im_height
    Data_cat = []
    Data_dog = []
    
    for i in range(1500) :
        filepath = "Project 2/Cat/" + str(i) +".jpg"
        im = Image.open(filepath)
        new_im =im.resize((im_width,im_height))
        new_im= new_im.convert('RGB')
        a = np.asarray(new_im)
        a=a/255
        Data_cat.append(a)
        
    for i in range(1500):
        filepath = "C:/Users/Akshay/Python_start/PAL/Project 2/dog/" + str(i) +".jpg"
        im = Image.open(filepath)
        new_im =im.resize((im_width,im_height), Image.NEAREST)
        new_im= new_im.convert("RGB")
        a = np.asarray(new_im)
        a=a/255
        #a.shape
        #print(a.shape)
        Data_dog.append(a)
       
    catx, catxcheck, caty, catycheck = train_test_split(np.array(Data_cat), np.array([1 for _ in range(1500)]), test_size = test_percent)
    dogx, dogxcheck, dogy, dogycheck = train_test_split(np.array(Data_dog), np.array([0 for _ in range(1500)]), test_size = test_percent)

    return catx, catxcheck, caty, catycheck, dogx, dogxcheck, dogy, dogycheck

def combine(test_percent,im_width, im_height):
    catx, catxcheck, caty, catycheck, dogx, dogxcheck, dogy, dogycheck = np.asarray(createdata(test_percent, im_width, im_height))
    x=np.concatenate((catx,dogx))
    y=np.concatenate((caty,dogy))
    xcheck=np.concatenate((catxcheck,dogxcheck))
    ycheck=np.concatenate((catycheck,dogycheck))
    #print(np.array(catx).shape,np.array(caty).shape,np.array(catxcheck).shape,np.array(catycheck).shape)
    return x,y,xcheck,ycheck
    
    
```

```python
x,y,xcheck,ycheck =combine(0.05,150,150)
x,y = shuffle(x,y)
```

```python
x_1=x-0.5
x_1 /= np.std(x_1, axis = 0)
```

```python
import os
os.environ["KERAS_BACKEND"] = "plaidml.keras.backend"
import keras
from keras.callbacks import ModelCheckpoint
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D,MaxPooling2D
from keras.utils import to_categorical
from keras.models import load_model
```

```python
y_cat = to_categorical(y)
y_catcheck = to_categorical(ycheck)
```

```python
model = Sequential()

model.add(Conv2D(16,kernel_size=(3,3),activation ='relu', input_shape = (150,150,3)))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(16,kernel_size=(3,3),activation ='relu'))#, input_shape = (100,100,3)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.2))

model.add(Conv2D(32,kernel_size=(3,3),activation ='relu'))#, input_shape = (100,100,3)))
model.add(MaxPooling2D(pool_size=(3,3)))
model.add(Dropout(0.5))


model.add(Flatten())

model.add(Dense(120,activation='relu'))
model.add(Dense(60,activation='relu'))
model.add(Dropout(0.1))

model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss ='binary_crossentropy', optimizer='Adam',metrics=['accuracy'])
```

```python
checkpoint = ModelCheckpoint(filepath='bestone_.hdf5', verbose=1, save_best_only=True)
ml = model.fit(x_1, y, epochs=50, validation_split = 0.1, callbacks=[checkpoint])
```

```python
model=load_model('bestone.hdf5')
model.load_weights('bestone.hdf5')
model.evaluate(xcheck,ycheck)
```

```python
import pandas as pd
import itertools
from sklearn.metrics import confusion_matrix, classification_report
y_pred = np.round(model.predict(xcheck))

cm = confusion_matrix(ycheck,y_pred)
plt.figure(figsize=(5,5))
plt.imshow(cm,interpolation = 'nearest', cmap='Blues')
plt.ylabel('True',size=20)
plt.xlabel('Predicted',size=20)
plt.xticks([0,1] ,['Dog',"Cat"], rotation=45)
plt.yticks([-0.5,1.5],['Dog',"Cat"])
thresh = cm.max() / 2.


for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    plt.text(j, i, cm[i, j], horizontalalignment='center', color='white' if cm[i, j] > thresh else 'black', size=25)

    plt.tight_layout()
```
